```python
from textblob import TextBlob
tweets = [
    "I love this product! It's amazing ",
    "This is the worst service ever ",
    "The experience was okay, nothing special."
]

print("Sentiment Analysis Results:\n")

for tweet in tweets:
    analysis = TextBlob(tweet)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    print(f"Tweet: {tweet}")
    print(f"Polarity: {polarity}")
    print(f"Sentiment: {sentiment}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
model = AutoModelForSeq2SeqLM.from_pretrained("sshleifer/distilbart-cnn-12-6")

text = """
Artificial Intelligence is transforming industries by enabling machines to learn from data.
It is widely used in healthcare, finance, education, and more.
AI helps in automation, decision-making, and improving efficiency.
However, it also raises concerns about privacy and job displacement.
"""

inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True)

summary_ids = model.generate(inputs["input_ids"], num_beams=4, max_length=50, min_length=20, early_stopping=True)

summary_text = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]

print("Original Text:\n", text)
print("\nSummary:\n", summary_text[0])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')

analyzer = SentimentIntensityAnalyzer()

tweet = "I love the new update, it's awesome!"

score = analyzer.polarity_scores(tweet)

print("Tweet:", tweet)
print("Scores:", score)

if score['compound'] >= 0.05:
    print("Sentiment: Positive")
elif score['compound'] <= -0.05:
    print("Sentiment: Negative")
else:
    print("Sentiment: Neutral")
```
