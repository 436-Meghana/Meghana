```python
import  nltk
from nltk.tokenize import sent_tokenize

```

```python
data = """Give Me Blood and I Will Give You Freedom" by Subhas Chandra Bose (1944): A pivotal rallying cry for the Indian National Army, inspiring the youth to fight for independence from British rule."""
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
data = data.lower()
sentences = sent_tokenize(data)

print(sentences)
```

```python
import re
# Ensure 'data' is a string defined in a previous cell before running this.
data = data.lower() # Convert entire string to lowercase
# Replace non-alphanumeric characters (excluding spaces) with a single space
data = re.sub(r'[^a-z0-9\s]', ' ', data)
# Replace multiple spaces with a single space and strip leading/trailing spaces
data = re.sub(r'\s+', ' ', data).strip()
```
