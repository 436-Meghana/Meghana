```python
import spacy
from nltk.stem import PorterStemmer
import nltk

# Download NLTK data (only first time)
nltk.download('punkt')

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple is looking at buying U.K. startup for $1 billion in 2023."

# Process text with spaCy
doc = nlp(text)

print("Original Text:")
print(text)

# 1. Tokenization
print("\n1. Tokenization:")
tokens = [token.text for token in doc]
print(tokens)

# 2. Stemming (using NLTK)
print("\n2. Stemming:")
ps = PorterStemmer()
stems = [ps.stem(token.text) for token in doc if token.is_alpha]
print(stems)

# 3. Lemmatization
print("\n3. Lemmatization:")
lemmas = [token.lemma_ for token in doc]
print(lemmas)

# 4. POS Tagging
print("\n4. POS Tagging:")
for token in doc:
    print(token.text, "->", token.pos_)

# 5. Named Entity Recognition (NER)
print("\n5. Named Entity Recognition:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)

# 6. Stop Word Removal
print("\n6. Stop Word Removal:")
filtered_tokens = [token.text for token in doc if not token.is_stop and token.is_alpha]
print(filtered_tokens)
```
