```python
# =====================================================
# COMPLETE ANN CODE (RUN THIS WHOLE BLOCK ONCE)
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.neural_network import MLPClassifier

# -------------------------
# 1. LOAD DATASET
# -------------------------

file_path = "/content/SpotifyAudioFeaturesApril2019.csv"
df = pd.read_csv(file_path)

# -------------------------
# 2. SELECT FEATURES
# -------------------------

features = ['danceability','energy','valence',
            'tempo','loudness','acousticness']

df = df.dropna()
X = df[features]

# Automatically select categorical target column
target_col = df.select_dtypes(include=['object']).columns[0]
y = df[target_col]

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# -------------------------
# 3. TRAIN TEST SPLIT
# -------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

# -------------------------
# 4. SCALING
# -------------------------

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =====================================================
# MODEL I (DEFAULT)
# =====================================================

model1 = MLPClassifier(
    hidden_layer_sizes=(64, 32),
    activation='relu',
    learning_rate_init=0.001,
    batch_size=32,
    max_iter=30,
    random_state=42
)

model1.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)

acc1 = accuracy_score(y_test, y_pred1)
prec1 = precision_score(y_test, y_pred1, average='weighted')
rec1 = recall_score(y_test, y_pred1, average='weighted')
f11 = f1_score(y_test, y_pred1, average='weighted')

print("\nModel I Results")
print("Accuracy:", acc1)
print("Precision:", prec1)
print("Recall:", rec1)
print("F1 Score:", f11)

# =====================================================
# MODEL II (WITH HYPERPARAMETER TUNING)
# =====================================================

param_grid = {
    'hidden_layer_sizes': [(128,64),(100,50),(64,32,16)],
    'activation': ['relu','tanh'],
    'learning_rate_init': [0.001,0.01],
    'batch_size': [32,64]
}

grid = GridSearchCV(
    MLPClassifier(max_iter=200, random_state=42),
    param_grid,
    cv=3,
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("\nBest Parameters:", grid.best_params_)

model2 = grid.best_estimator_

y_pred2 = model2.predict(X_test)

acc2 = accuracy_score(y_test, y_pred2)
prec2 = precision_score(y_test, y_pred2, average='weighted')
rec2 = recall_score(y_test, y_pred2, average='weighted')
f12 = f1_score(y_test, y_pred2, average='weighted')

print("\nModel II Results")
print("Accuracy:", acc2)
print("Precision:", prec2)
print("Recall:", rec2)
print("F1 Score:", f12)

# =====================================================
# CONFUSION MATRIX
# =====================================================

cm = confusion_matrix(y_test, y_pred2)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
plt.figure()
disp.plot()
plt.title("Confusion Matrix - Tuned Model")
plt.show()
```
