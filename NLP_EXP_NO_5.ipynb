```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
para = """Bluetooth is a wireless communication technology that allows devices to connect and exchange data over short distances using radio waves.
It is commonly used to link smartphones with headphones, speakers, keyboards, and other electronic devices without the need for cables. Bluetooth operates on low power, making it efficient for portable gadgets and helping to conserve battery life.
 The technology was designed to be secure and reliable, using pairing methods and encryption to protect data during transmission."""
```

```python
sentence = nltk.sent_tokenize(para)
```

```python
sentence
```

```python
sentence
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(sentence)):
  words = nltk.word_tokenize(sentence[i])
  words = [stemmer.stem(word) for word in words]
  sentence[i] = ' '.join(words)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
para1 = """Bluetooth is a wireless communication technology that allows devices to connect and exchange data over short distances using radio waves.
It is commonly used to link smartphones with headphones, speakers, keyboards, and other electronic devices without the need for cables. Bluetooth operates on low power, making it efficient for portable gadgets and helping to conserve battery life.
The technology was designed to be secure and reliable, using pairing methods and encryption to protect data during transmission."""
```

```python
sentence1 = nltk.sent_tokenize(para1)
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentence1)):
  words = nltk.word_tokenize(sentence1[i])
  newwords = [lemmatizer.lemmatize(word) for word in words]
  sentence1[i] =' '.join(newwords)
```
